 Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot. 

In seaborn Library we can use those following fives plots :-  
1. scatterplot
2. displot
3. relplot
4. catplot
5. jointplot

```python
import numpy as np
import seaborn as sns
# Load the fmri dataset
taxis = sns.load_dataset('taxis')
taxis.head()
```

Scatterplot:

Use: Scatterplots are used to visualize the relationship between two continuous variables. Each point on the plot represents a data point, and the position of the point on the X and Y axes corresponds to the values of the two variables. Scatterplots are useful for identifying patterns, correlations, or clusters in the data.

```python
sns.scatterplot(x = taxis.fare, y = taxis.passengers, hue=taxis.color ) # here we are plot scatter plot fair and passengers relationship and contect with taxi colour
```

Displot (Distribution Plot):

Use: Displots are used to visualize the distribution of a single variable. They show the frequency or probability distribution of a continuous variable. Displots can help us understand the central tendency, spread, and skewness of the data.

```python
sns.displot(x = np.log10(taxis.fare), hue=taxis.payment ) 
```

Relplot (Relational Plot):

Use: Relational plots are used to visualize the relationship between two variables, often with a categorical variable determining the color of the points. They are exploring how two variables interact or change concerning a third variable.

```python
sns.relplot(x =taxis.fare, y = taxis.distance , hue = taxis.color, style=taxis.passengers  )
```

Catplot (Categorical Plot):

Use: Catplots are used to visualize the relationship between one categorical variable and one continuous or categorical variable. They can show summary statistics such as means and confidence intervals for the data within each category. Catplots are useful for comparing groups or categories.

```python
sns.catplot(x =taxis.pickup_borough, y = (taxis.tip)) 
```

Jointplot:

Use: Jointplots are used to visualize the relationship between two variables along with their univariate distributions. They combine scatterplots, histograms, and density plots to provide a comprehensive view of the data. Jointplots are helpful for understanding both the relationship and the marginal distributions of variables.

```python
sns.jointplot(x = taxis.fare, y = taxis.tip, hue =taxis.pickup_borough )
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =  "timepoint" and y = "signal" for different events and regions.  
Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create two line plots 
plt.figure(figsize=(12, 5))

# line plot for 'region'
plt.subplot(1, 2, 1)
sns.lineplot(x =  "timepoint", y = "signal",data = fmri , hue = "region" )

# line plot for 'event'
plt.subplot(1, 2, 2)
sns.lineplot(x =  "timepoint", y = "signal",data = fmri , hue = "event" )
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =  'pclass', y = 'age' and y = 'fare'. 
Note:  pclass, age, and fare are columns in the titanic dataset. 

```python

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create two box plots 
plt.figure(figsize=(12, 5))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column  of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset or create a DataFrame
data = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(data, hue='species')

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix of passenger counts for each month and year
flights_pivot = flights.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count (in thousands) by Month and Year")

plt.show()
```

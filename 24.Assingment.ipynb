    Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot  is a type of data visualization that is used to display the relationship between two continuous variables. In a scatter plot, each data point is represented as a single point on the graph, with one variable on the x-axis and the other variable on the y-axis. The position of each point is determined by its values of the two variables being compared. 

```python
import matplotlib.pyplot as plt
import numpy as np 
np.random.seed(3) 
x =  3+ np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x)) 
plt.scatter(x,y)  
plt.xlabel('This is X lable')
plt.ylabel('This is Y lable')
plt.title('Scatter Plot')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data: 

    import numpy as np 
    For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])
    For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70]) 
    For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60]) 
    For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The subplot() function in Matplotlib is used to create multiple subplots (i.e., multiple plots within a single figure) so that we can display different plots side by side or in a grid. It helps in organizing and visualizing data efficiently.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots in a 2x2 grid
plt.figure(figsize=(12,8))
# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

# Adjust spacing between subplots
plt.tight_layout(pad=1.8)

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.  
    
    import numpy as np 
    company = np.array(["Apple", "Microsoft", "Google", "AMD"]) 
    profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a common data visualization tool used to represent categorical data. It displays data with rectangular bars, where the length or height of each bar is proportional to the value it represents. Bar plots are widely used for several purposes:

1. Comparing Categories:  Bar plots are excellent for comparing the sizes or counts of different categories or groups. Each category is represented by a separate bar, making it easy to see which categories have larger or smaller values.

2. Showing Distribution: Bar plots can display the distribution of data within each category. For example, we can use them to show the frequency of different responses to a survey question.

3. Tracking Changes Over Time: In a time-series bar plot, categories are plotted along the x-axis, and the height of the bars represents values at different time points. This allows us to track changes in categories over time.

4. Visualizing Ranking: Bar plots can help visualize rankings or ordering of categories. We can order the bars from highest to lowest or vice versa to see which categories are the most or least significant.

5. Categorical Comparison: Bar plots are especially useful when we have categorical data, such as types of fruits, cities, or survey responses. They provide a clear visual representation of how different categories compare.

6. Presenting Data Clearly: Bar plots are straightforward and easy to interpret, making them a popular choice for presenting data to a wide audience.


```python
import matplotlib.pyplot as plt
import numpy as np 
company = np.array(["Apple", "Microsoft", "Google", "AMD"]) 
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title("Vertical Bar")
plt.show()
```

```python
plt.barh(company,profit)
plt.ylabel("Company Name")
plt.xlabel("Profit")
plt.title("Horizontal Bar")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. It is commonly used to visualize the summary statistics and distribution of a set of data points. Here's how a box plot works and why it's used:

Components of a Box Plot:

1. Box: The box represents the interquartile range (IQR), which is the range between the first quartile (Q1, 25th percentile) and the third quartile (Q3, 75th percentile). It contains the middle 50% of the data.

2. Line Inside the Box (Median): The line inside the box represents the median (also known as Q2 or the 50th percentile) of the data. The median is the middle value when the data is ordered from smallest to largest.

3. Whiskers: The whiskers extend from the edges of the box to the minimum and maximum values within a certain range. These values are often calculated using the interquartile range (IQR) method or other methods.

4. Outliers: Data points that fall significantly below the first quartile (Q1) or above the third quartile (Q3) are considered outliers and are typically shown as individual points beyond the whiskers.

Why Box Plots Are Used:

1. Visualizing Data Distribution: Box plots provide a clear and concise way to visualize the distribution of data, including its central tendency, spread, and presence of outliers.

2. Comparing Distributions: Box plots are useful for comparing the distributions of multiple datasets or groups. They allow us to quickly assess differences in medians, spreads, and the presence of outliers.

3. Identifying Outliers: Box plots make it easy to identify outliers in a dataset, which can be important for data quality assessment and anomaly detection.

4. Summary Statistics: Box plots display summary statistics, including the median, quartiles, and range, in a compact format, making it easy to grasp essential information about the data.

5. Robustness: Box plots are robust to extreme values and outliers, making them a reliable choice for summarizing data with potential outliers.

6. Data Exploration: In data exploration and data analysis, box plots are a valuable tool for getting an initial sense of the data's characteristics before more in-depth analysis.


```python
box1 = np.random.normal(100, 10, 200) 
box2 = np.random.normal(90, 20, 200) 
plt.boxplot([box1,box2])    
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.title('Box Plot')
```
